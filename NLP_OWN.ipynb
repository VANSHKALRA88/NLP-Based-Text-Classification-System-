```python
df = pd.read_csv('train.txt.zip', sep=';', header=None, names=['text','emotion'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
unique_emotion=df['emotion'].unique()
emotion_number={}
i=0
for emo in unique_emotion:
    emotion_number[emo]=i
    i+=1

df['emotion']=df['emotion'].map(emotion_number)
```

```python
df
```

```python
df['text']=df['text'].apply(lambda x: x.lower())
```

```python
import string
def remove_punctuation(x):
    return x.translate(str.maketrans('','',string.punctuation))
```

```python
df['text']=df['text'].apply(remove_punctuation)
```

```python
def remove_numbers(txt):
  new=''
  for i in txt:
    if not i.isdigit():
      new+=i
  return new
df['text']=df['text'].apply(remove_numbers)
```

```python
def remove_emojis(txt):
  new=''
  for i in txt:
    if i.isascii():
      new+=i
  return new

df['text']=df['text'].apply(remove_emojis)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
```

```python
df.loc[1]['text']
```

```python
def remove(txt):
  words=word_tokenize(txt)
  cleaned_text=[]
  for i in words:
    if not i in stop_words:
      cleaned_text.append(i)
  return ' '.join(cleaned_text)
```

```python
df['text']=df['text'].apply(remove)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     df['text'], df['emotion'], test_size=0.20, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer
```

```python
bow_vectorizer=CountVectorizer()
```

```python
x_train_bow=bow_vectorizer.fit_transform(X_train)
x_train_bow
```

```python
x_test_bow=bow_vectorizer.transform(X_test)
x_test_bow
```

```python
from sklearn.naive_bayes import MultinomialNB
```

FOR CLASSIFICATION OR FREQUENCY AND STOPWORDS THEN WE USE MULTINOMIAL NAIVE BIAS

```python
from sklearn.metrics import accuracy_score
```

```python
nb_model=MultinomialNB()
nb_model.fit(x_train_bow,y_train)
```

```python
pred_nb=nb_model.predict(x_test_bow)
```

```python
print(accuracy_score(y_test,pred_nb))
```

```python
tfidf_vectorizer=TfidfVectorizer()
```

```python
x_train_tfidf=tfidf_vectorizer.fit_transform(X_train)
x_test_tfidf=tfidf_vectorizer.transform(X_test)
```

```python
nb2_model=MultinomialNB()
```

```python
nb2_model.fit(x_train_tfidf,y_train)
```

```python
y_pred=nb2_model.predict(x_test_tfidf)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model=LogisticRegression(max_iter=1000)
```

```python
logistic_model.fit(x_train_tfidf,y_train)
```

```python
log_pred=logistic_model.predict(x_test_tfidf)
```
